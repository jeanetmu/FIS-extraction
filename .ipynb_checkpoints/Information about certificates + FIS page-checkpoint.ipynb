# Information we have gotten about the different courses so far 

- __Opposite of what we expected, the Date of Issue does not always correspond with the homoglation number listed on the website :')__
<br> <br>
Example: "NOR_Bodo_20_51-05_2-5.png" where the original registration number for the certificate was 15/04.05/2,5. We can see that the Date of Issue was indeed in 2015, but this doesn't show in the name of the certificate-file

We can double-check that the certificate was not crawled with the wrong homoglation number in the filename by checking with the link available on the FIS homepage: https://medias3.fis-ski.com/pdf/homologations/CC/NOR/NOR_Bodo_20_51-05_2-5.pdf
<br> Also in this link can we see that the filename in the path is the same as the saved filename for the certificate
***

- __File size does not disclose whether the certificate is a scanned document or a digital copy__

We can see this by manually comparing the different pdfs. While some the small sized images with a size *>=633kB* were in fact scanned documents, I also found scanned files with a size of *2,1MB*, making them the biggest files in the folder. Seeing that we at this point seem to be missing some image files in "NOR-pdfs/", I will not assume that 633kb is a border that remains true for the so far unseen certificates. 
<br> The digital (and searchable) certificates also differed, within the range *1,5MB-633kB*
